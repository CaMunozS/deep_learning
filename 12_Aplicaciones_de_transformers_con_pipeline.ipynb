## Importamos librerías

Usaremos [`pipeline()`](https://huggingface.co/docs/transformers/en/pipeline_tutorial) de Hugging Face.

`pipeline()` es una interfaz de alto nivel que agrupa:

- el modelo preentrenado,

````python
from transformers import pipeline
import pandas as pd
````

## Primera tarea: análisis de sentimiento

Una de las tareas más clásicas en NLP es clasificar el sentimiento de un texto.

En Hugging Face, esta tarea está disponible directamente como:

- `sentiment-analysis`

````python
sentiment_classifier = pipeline("sentiment-analysis")
````

## ¿Qué significa el mensaje que apareció?

Cuando ejecutamos:

```python
sentiment_classifier = pipeline("sentiment-analysis")



Hugging Face nos informa que, como no especificamos manualmente un modelo, eligió uno por defecto para esta tarea.
En este caso, el pipeline indica algo como:
`No model was supplied, defaulted to distilbert/distilbert-base-uncased-finetuned-sst-2-english ...`
Eso significa que el sistema seleccionó automáticamente un modelo ya entrenado para análisis de sentimiento.

## ¿Qué es BERT?

**BERT** es una arquitectura de transformers muy influyente en procesamiento de lenguaje natural.

Fue diseñada principalmente para aprender representaciones de texto y luego adaptarse a tareas como:

- clasificación,
- pregunta-respuesta,
- análisis de sentimiento,
- entre otras.

En términos simples, BERT es uno de los modelos que ayudó a consolidar el uso masivo de transformers en NLP.

## ¿Qué es DistilBERT?

**DistilBERT** es una versión más pequeña y más rápida de BERT.

La idea es conservar gran parte del desempeño de BERT, pero usando menos parámetros y menos costo computacional.

Esto se logra mediante un proceso llamado **knowledge distillation** o destilación de conocimiento.

En términos prácticos:

- BERT suele ser más grande,
- DistilBERT suele ser más liviano y más rápido,
- por eso DistilBERT es una opción muy conveniente para demos, notebooks y aplicaciones ligeras.

## Descomponiendo el nombre del modelo

El nombre completo es:

`distilbert-base-uncased-finetuned-sst-2-english`

Podemos leerlo por partes:

- **distilbert**: pertenece a la familia DistilBERT.
- **base**: indica una versión base del modelo, no una versión gigante.
- **uncased**: el modelo no distingue entre mayúsculas y minúsculas.
- **finetuned**: ya fue ajustado para una tarea específica.
- **sst-2**: fue ajustado usando el dataset Stanford Sentiment Treebank 2.
- **english**: está pensado para texto en inglés.

## ¿Qué significa “fine-tuned on SST-2”?

Primero, el modelo fue **preentrenado** de manera general sobre grandes cantidades de texto.

Después, fue **fine-tuned** o ajustado finamente en una tarea concreta:
clasificación de sentimiento.

Para eso se usó **SST-2**, un dataset clásico de análisis de sentimiento en inglés.

Por eso este modelo ya viene especialmente preparado para distinguir entre textos positivos y negativos.

## ¿Qué significa `revision 714eb0f`?

La palabra **revision** se refiere a una versión concreta del repositorio del modelo en Hugging Face.

Esa cadena corta (`714eb0f`) funciona como un identificador de revisión o commit.

La idea es simple:

- el nombre del modelo dice **qué modelo** estamos usando,
- la revisión indica **qué versión exacta** de ese modelo o repositorio quedó seleccionada.

Esto ayuda a la reproducibilidad, porque dos personas pueden cargar exactamente la misma versión del modelo.

## Idea importante para quedarse

Cuando usamos:

```python
pipeline("sentiment-analysis")

no estamos creando un modelo desde cero.
Estamos usando un sistema que automáticamente selecciona:


*   un modelo preentrenado,
*   su tokenizer correspondiente,
*   y una configuración lista para inferencia.

Eso es justamente parte de la potencia del ecosistema moderno:
podemos aplicar modelos avanzados con muy poco código.

````python
texts = [
    "I loved this movie. The acting was excellent.",
    "This film was boring and disappointing.",
    "The notebook was clear and useful."
]

results = sentiment_classifier(texts)
results
````

## ¿Qué observamos?

El pipeline devuelve, para cada texto:

- una etiqueta predicha,
- un score de confianza.

Esto nos permite usar un modelo moderno sin tener que tokenizar manualmente ni definir la arquitectura capa por capa.

````python
df_results = pd.DataFrame({
    "text": texts,
    "prediction": [r["label"] for r in results],
    "score": [r["score"] for r in results]
})

df_results
````

## Probemos con ejemplos nuevos

Ahora podemos evaluar textos creados por nosotros.

````python
new_texts = [
    "I really enjoyed the class and learned a lot.",
    "The explanation was confusing and too rushed.",
    "The movie had good visuals, but the story was weak."
]

new_results = sentiment_classifier(new_texts)

pd.DataFrame({
    "text": new_texts,
    "prediction": [r["label"] for r in new_results],
    "score": [r["score"] for r in new_results]
})
````

## Segunda tarea: zero-shot classification

Otra tarea interesante es la **zero-shot classification**.

Aquí el modelo intenta clasificar un texto usando etiquetas que nosotros proponemos, incluso si no entrenamos el modelo específicamente para ese conjunto de clases.

````python
zero_shot_classifier = pipeline("zero-shot-classification")
````

## ¿Qué es `facebook/bart-large-mnli`?

En este caso, Hugging Face seleccionó por defecto el modelo:

`facebook/bart-large-mnli`

Podemos leerlo por partes:

- **facebook**: indica la organización o autor que publicó originalmente el modelo en Hugging Face.
- **bart**: corresponde a la familia **BART**, un modelo transformer de tipo encoder-decoder.
- **large**: indica una versión grande del modelo.
- **mnli**: significa que fue ajustado usando el dataset **Multi-Genre Natural Language Inference (MNLI)**.

La idea importante es esta:

este modelo no fue entrenado directamente como un clasificador tradicional de temas o categorías, sino como un modelo de **inferencia en lenguaje natural**.

Eso permite reutilizarlo para **zero-shot classification**.

En términos simples, el modelo evalúa si un texto es compatible con una etiqueta candidata formulada como una hipótesis, por ejemplo:

- texto: `"This course session introduces transformers and language models."`
- hipótesis: `"This text is about education."`

Luego estima qué tan bien esa hipótesis está apoyada por el texto.

Por eso `facebook/bart-large-mnli` funciona tan bien como modelo base para clasificación zero-shot.

````python
candidate_labels = ["education", "entertainment", "sports", "technology"]

text = "This course session introduces transformers, attention, and language models."

zero_shot_result = zero_shot_classifier(text, candidate_labels)
zero_shot_result
````

## Interpretación

El modelo compara el texto con las etiquetas candidatas y estima cuál parece más adecuada.

Esto muestra una capacidad moderna muy distinta a los clasificadores tradicionales entrenados desde cero sobre un conjunto fijo de clases.

## Y en español?

````python
from transformers import pipeline

zero_shot_classifier = pipeline(
    "zero-shot-classification",
    model="Recognai/bert-base-spanish-wwm-cased-xnli"
)

texto = "Esta sesión introduce transformers, atención y modelos de lenguaje."
labels = ["educación", "deporte", "tecnología", "entretenimiento"]

zero_shot_classifier(texto, labels)
````

## Tercera tarea: question answering

También podemos usar transformers preentrenados para responder preguntas a partir de un contexto dado.

````python
qa_pipeline = pipeline("question-answering")
````

## ¿Qué es `distilbert/distilbert-base-cased-distilled-squad`?

Para la tarea de **question answering**, Hugging Face seleccionó por defecto el modelo:

`distilbert/distilbert-base-cased-distilled-squad`

Podemos leerlo por partes:

- **distilbert**: pertenece a la familia **DistilBERT**, una versión más liviana y rápida de BERT.
- **distilbert-base**: corresponde a una versión base del modelo.
- **cased**: el modelo **sí distingue mayúsculas y minúsculas**.
- **distilled**: fue obtenido mediante **knowledge distillation**, es decir, conservando gran parte del conocimiento de un modelo mayor en una versión más compacta.
- **squad**: fue ajustado para la tarea de pregunta-respuesta usando el dataset **SQuAD**.

La idea importante es esta:

este modelo ya fue preparado para responder preguntas a partir de un **contexto dado**.

Por eso, cuando usamos el pipeline de `question-answering`, el modelo intenta encontrar dentro del texto el fragmento que mejor responde la pregunta.

````python
context = """
Transformers are deep learning architectures based on attention mechanisms.
They became highly influential in natural language processing and later extended to vision, audio, and multimodal tasks.
"""

question = "What is the main mechanism behind transformers?"

qa_result = qa_pipeline(question=question, context=context)
qa_result
````

## Comentario importante

En este caso, el modelo no está “inventando” libremente una respuesta:
está intentando encontrar una respuesta dentro del contexto entregado.

Esto es distinto de la generación abierta de texto.

## Y en español?

````python
from transformers import pipeline

qa_pipeline = pipeline(
    "question-answering",
    model="mrm8488/bert-multi-cased-finetuned-xquadv1",
    tokenizer="mrm8488/bert-multi-cased-finetuned-xquadv1"
)

contexto = """
Los transformers son arquitecturas de deep learning basadas en mecanismos de atención.
Primero fueron muy influyentes en procesamiento de lenguaje natural y luego también se extendieron a visión.
"""

pregunta = "¿En qué mecanismo se basan los transformers?"

qa_pipeline(question=pregunta, context=contexto)
````

## Cuarta tarea: generación de texto

Muchos modelos modernos de lenguaje también se usan para continuar o generar texto.

````python
text_generator = pipeline("text-generation", model="distilgpt2")
````

## ¿Qué es `distilbert/distilgpt2`?

Para la tarea de **generación de texto**, en este ejemplo usamos el modelo:

`distilbert/distilgpt2`

Podemos leerlo por partes:

- **distilbert**: indica la organización o comunidad que publica este modelo en Hugging Face.
- **distilgpt2**: corresponde a **DistilGPT2**, una versión más liviana de GPT-2.

La idea importante es esta:

- **GPT-2** es un modelo transformer orientado a **generación de texto**.
- **DistilGPT2** es una versión más pequeña y más rápida, obtenida mediante **knowledge distillation**.
- Por eso resulta útil para demos y notebooks donde queremos generar texto sin usar un modelo demasiado pesado.

A diferencia de los modelos usados antes para clasificación o pregunta-respuesta, aquí el objetivo no es asignar una etiqueta ni buscar una respuesta en un contexto, sino **continuar texto** generando nuevos tokens.

En otras palabras, este modelo toma un texto inicial y predice cómo podría seguir.

````python
generated = text_generator(
    "Deep learning models based on attention are useful because",
    max_new_tokens=40,
    num_return_sequences=1
)

generated
````

## Y en español?

````python
generator = pipeline(
    "text-generation",
    model="DeepESP/gpt2-spanish"
)

generator("Los transformers son importantes porque", max_new_tokens=40)
````
